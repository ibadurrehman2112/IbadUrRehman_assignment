```python
my_data = [2.62521e+10,1.54502e+10,6.24013e+08]
my_labels = 'Confirmed','Recovered', 'Deaths'
my_explode = (0,0,1.5)
my_colors = ['lightsteelblue','lightpink','grey']
fig1, ax1 = plt.subplots(figsize=(13,8))
plt.pie(my_data,labels=my_labels,autopct= '%1.1f%%', startangle = 15 , shadow=True, colors=my_colors, explode=my_explode)
plt.axis('equal')
plt.show()
#PIE chart visualization 
```

```python
df[['Country/Region','Confirmed']].groupby(['Country/Region']).mean().sort_values(by='Confirmed').plot.bar(color='purple')
plt.show()
#We can also use the groupby function to sort values in an ascending order based on the x-axis,y-axis, and its keys
#Below we get a clear picture of countries in an increasing order based on their confirmed level of cases.
#Turkey has high level of confirmed cases compared to other countries.
#Also some countries data are not updated or shared, we need to use updated data for new actual graphs
```

```python
df.columns
```

```python
df[['Country/Region','Recovered']].groupby(['Country/Region']).mean().sort_values(by='Recovered').plot.bar(color='purple')
plt.show()
#We can also use the groupby function to sort values in an ascending order based on the x-axis,y-axis, and its keys
#Below we get a clear picture of countries in an increasing order based on their Recovered level of cases.
#Turkey has high level of Recovered cases compared to other countries.
#Also some countries data are not updated or shared, we need to use updated data for new actual graphs
```

```python
df.columns
```

```python
df[['Country/Region','Deaths']].groupby(['Country/Region']).mean().sort_values(by='Deaths').plot.bar(color='purple')
plt.show()
#We can also use the groupby function to sort values in an ascending order based on the x-axis,y-axis, and its keys
#Below we get a clear picture of countries in an increasing order based on their Deaths level of cases.
#Iran has high level of Deaths cases compared to other countries.
#Also some countries data are not updated or shared, we need to use updated data for new actual graphs
```
