```python
# import libarires 
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Normal distribution 
def pdf(x):
    mean = np.mean(x)
    std = np.std(x)
    y_out = 1/(std*np.sqrt(2*np.pi))*np.exp(-(x-mean)**2 / (2*std**2))
    return y_out
# to genreate an array of x 
x = np.arange(-2, 2, 0.1)
y = pdf(x)

# ploting the normal curve / bell curve or guassian distribution 

plt.style.use('seaborn')
plt.figure(figsize=(5,5))

plt.plot(x, y, color = 'blue', linestyle = 'dashed')

plt.scatter(x,y, marker='o', s=25, color='red')
```
