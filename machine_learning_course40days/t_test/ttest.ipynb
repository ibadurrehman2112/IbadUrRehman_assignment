## Student's ttest
1. one-sample t-test
2. Two-sample t-test
   1. Un-paired or Independent t-test
   2. Paired or relational/dependent t-test

## One-sample student's t test 
Test a sample with a known standard value. **Assumptions**
- Observations in sample is independent and identically distributed.
- Observations in sample is normally distributed.

**Interpretation**
>**H0:** The mean of the samples is equal to the known value.\
>**H1:** The mean of the samples is unequal to the known value.

```python
# One sample t-test
import seaborn as sns
import pandas as pd 
from scipy.stats import ttest_1samp

# Load dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df1 = df[['sex', 'age', 'fare']]
df1.head()
```

```python
df1.describe()
```

```python
# Check the age and compare with a known value of 45 years
ttest_1samp(df1['fare'], 50)
stat, p = ttest_1samp(df1['fare'], 50)
print('stat=%.3f, p=%.3f' % (stat, p))
#adding conditional argument for ease 
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```

## Two-sample student's t test 
**Independent student's ttest**
**Assumptions**
Test a sample with a known standard value. 
- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.

**Interpretation**
>**H0:** The means of the samples are equal.\
>**H1:** The means of the samples are unequal.

```python
# We will compare age and fare of male vs female passengers

# splitting data sets

df_male = df1.loc[df1['sex']=='male']
df_female= df1.loc[df1['sex']=='female']

# library
from scipy.stats import ttest_ind

stat, p = ttest_ind(df_male['age'], df_female['age'])

print('stat=%.3f, p=%.3f' % (stat, p))
#adding conditional argument for ease 
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```

## Paired student's t test 
**Independent student's ttest**

Tests whether the mean of two paired smaple are significantly different.
**Assumptions**

- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.
- Observations across each sample are paired.

**Interpretation**
>**H0:** The means of the samples are equal.\
>**H1:** The means of the samples are unequal.

```python
df.head()
```

```python
# Select only male data 
df_m = df.loc[df['sex']=='male']
df_m.head()
```

```python
# Select only two classes 
df_male_first = df_m.loc[df_m['class']=='First']
df_male_second = df_m.loc[df_m['class']=='Second']
df_male_third = df_m.loc[df_m['class']=='Third']
```

```python
# Check our data 
df_male_third.head()
```

```python
df_1st = df_male_first.sample(n=100)
df_2nd = df_male_second.sample(n=100)
print("The number of instances in 1st class are =", df_1st.shape)
print("The number of instances in 1st class are =", df_2nd.shape)
```

```python
from scipy.stats import ttest_rel
# apply test to compare class-1 and class 3
stat,  p = ttest_rel(df_1st['age'], df_2nd['age'])

print('stat=%.3f, p=%.3f' % (stat, p))
#adding conditional argument for ease 
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distribution')
```
